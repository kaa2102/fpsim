```python
#Import Libraries
import sciris as sc
import fpsim as fp
```

```python
# Set options
do_plot = True
pars = fp.pars(location='senegal')
pars['n_agents'] = 2000 # Small population size
pars['end_year'] = 2020 # 1961 - 2020 is the normal date range
pars['exposure_correction'] = 1.0 # Overall scale factor on probability of becoming pregnant
```

```python
#Run Simulation
sc.tic()
sim = fp.Sim(pars=pars)
sim.run()
```

```python
#Do Plot is already set to True
if do_plot:
    sim.plot()

sc.toc()
print('Done.')
```
